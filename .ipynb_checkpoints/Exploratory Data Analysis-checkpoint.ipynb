### Data Cleaning

```python
# Make a copy of the original DataFrame
df = data.copy()
```

#### 1. Handling Missing Values

```python
# Imputing categorical columns with mode for missing values
for col in df.select_dtypes(include = 'object').columns:
    df[col].fillna(df[col].mode()[0], inplace = True)
```

```python
# Impute numerical columns with the median for missing values
for col in df.select_dtypes(include = np.number).columns:
    df[col].fillna(df[col].median(), inplace = True)
```

```python
# Recheck if any missing values
df.isnull().sum()
```

The null values have now been handled

#### 2.Data Type Convertion

```python
# Convert ID into object type
df['ID'] = df['ID'].astype('O')
```

### Data Reduction

```python
# Drop Year column since it is a constant column
df.drop(columns = ['year'], axis = 1, inplace = True)
```

```python
# Save the dataset
df.to_csv('Clean_data.csv', index = False)
```

#### Gender

Q. Which gender default more?

```python
# Group by gender
gender_df = df.groupby('Gender').count()['Status'].sort_values(ascending = True)
x = gender_df.index
y = gender_df.values

# Plot a pie chart
plt.pie(y, labels = x, autopct = '%1.1f%%', startangle = 90, explode = [0, 0, 0, 0.1], shadow = True)

# Add labels
plt.title("Gender Distribution")

# Show plot
plt.show()
```

Findings:
- Male gender default more

#### Age

Q. Which age group default most?

```python
# create a DataFrame for the people who default only
default_df = df[df['Status'] == 1]

# Do a gender count plot
sns.countplot(data = default_df, x = 'age', color = 'green', edgecolor = 'black')

# Show the plot
plt.show()
```

Findings:
- Male gender default more 

#### Bivariate Analysis

#### Loan Amount vs Interest Rate

```python
# Plot a scatter plot
sns.scatterplot(data = df, x = 'loan_amount', y = 'rate_of_interest', hue = 'Status')

# Show the plot
plt.show()
```

```python
# Plot a scatter plot
sns.lmplot(data = df, x = 'loan_amount', y = 'rate_of_interest', scatter_kws = {'s' : 50}, line_kws = {'color' : 'red'})

# Show the plot
plt.show()
```

Findings:

#### Income vs Loan Amount

```python
# Plot a scatter plot
sns.scatterplot(data = df, x = 'income', y = 'loan_amount', hue = 'Status')

# Show the plot
plt.show()
```

```python
# Plot a scatter plot with a trend line
sns.lmplot(data=df, x='income', y='loan_amount', hue='Status', scatter_kws={'s': 50}, line_kws={'color': 'red'})

# Show the plot
plt.show()
```

Findings:
- It is clear that as income increase the loan amount also decrease.
- People with low income tend to default more than people with high income.

#### Credit Score vs Loan Amount

```python
# Plot the relationship and add a regression line
sns.lmplot(data = df, x = 'loan_amount', y = 'Credit_Score', hue = 'Status', scatter_kws = {'s' : 40})

# Show
plt.show()
```

Findings:
- It is clear that People with low loan_amount default more.

#### Property Value vs Loan Amount

```python
sns.lmplot(data = df, x = 'property_value', y = 'loan_amount', hue = 'Status', scatter_kws = {'s' : 40})

# show the plot
plt.show()
```

Findings:
- It is clear that as property value increases, loan amount also increases.
- Also People with a lower property value tend to default more.

#### Debt to income Ratio vs Income

```python
# Plot the scatter plot
sns.lmplot(data = df, x = 'income', y = 'dtir1', scatter_kws = {'s' : 40}, line_kws = {'color' : 'red'})

# Show the plot
plt.show()
```

### Multivariate Analysis

##### Correlation

```python
# Select important numeric columns
numeric_df = df.select_dtypes(include = np.number)

# Drop status colum
numeric_df = numeric_df.drop(columns = ['Status'])

# Create a correlation matrix
corr_matrix = numeric_df.corr()

# Display the numeric column
corr_matrix
```

```python
# Create a figure
plt.figure(figsize = (10,6))

# Visualize using a heatmap
sns.heatmap(corr_matrix, annot = True, cmap = 'coolwarm')

# Plot the heatmap
plt.show()
```

Key Findings:
- As propert value increases, the loan amount also increases
- As income Increases, Loan amount also Increases
